```python
def model_for_fit(t, y, K_ser, k_sRNA, k_dRNA, k_vp, K_RNA, k_t):
    X, DNA, RNA, SRIP = y
    
    dXdt = P * X * (X_max - X) / X_max * Ser / (Ser + K_ser) - k_t * SRIP
    dDNAdt = -k_DNA * DNA
    dRNAdt = k_sRNA * DNA - k_dRNA * RNA
    dSRIPdt = k_vp * RNA / (RNA + K_RNA) * X

    return [dXdt, dDNAdt, dRNAdt, dSRIPdt]

params_fix = [
    # P,             X_max, Ser,    k_DNA,        
    [math.log(2)/24, 1.3e6, 4.0e-4, math.log(2)/24]
    ]

initial_fitting_param = [
    # 初期値は論文の値を参照
    #K_Ser,  k_sRNA, k_dRNA, k_vp, K_RNA, k_t
    [1.0e-5, 3e-2,   3e-2,   1e-2, 2e+3,  1e-2]
    #fix,    fix,    fix,    TBD,  fix,   TBD
    ]

bounds = [(1.0e-8, 1.0e-2), (3.0e-5, 3.0e+1), (3.0e-5, 3.0e+1), (1.0e-5, 1.0e+1), (2.0e+0, 2.0e+6), (1.0e-5, 1.0e+1)]

P, X_max, Ser, k_DNA = params_fix[0]

X_0 = 2e5
DNA_0 = 2.0e11
RNA_0 = 0.0
SRIP_0 = 0.0
initial_value = [X_0, DNA_0 / X_0 * 0.8, RNA_0, SRIP_0] # X, DNA, RNA, SRIP
print(initial_value)
```

```python
triggered_t = None  # イベントがトリガされた時の時刻tを保存

def stop_event(t, y, *params):
    global triggered_t
    X, DNA, RNA, SRIP = y
    if X < 0 or DNA < 0 or RNA < 0 or SRIP < 0:
        triggered_t = t  # イベントがトリガされた時の時刻を保存
        return 0
    return 1

stop_event.terminal = True

def fit_function(t, K_ser, k_sRNA, k_dRNA, k_vp, K_RNA, k_t):
    global triggered_t

    if np.isscalar(t):
        t_values = [t]
    else:
        t_values = t

    results = []

    for t_val in t_values:
        if triggered_t is not None and t_val >= triggered_t:
            results.append(last_srip_value)
            continue

        sol = solve_ivp(
            model_for_fit, 
            [t_min, t_end], 
            initial_value, 
            args=(K_ser, k_sRNA, k_dRNA, k_vp, K_RNA, k_t),
            t_eval=[t_val], 
            method='RK45',
            events=stop_event
        )

        if triggered_t is not None and t_val >= triggered_t:
            results.append(last_srip_value)
        else:
            results.append(sol.y[3, -1])

    if np.isscalar(t):
        return results[0]
    else:
        return np.array(results)
```
